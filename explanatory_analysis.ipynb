```python
import os
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("Toyota.csv", index_col=0, na_values=["??", "????"])
```

```python
copy = cars.copy()
copy.head()
```

```python
pd.crosstab(index=copy["FuelType"], columns="count", dropna=True)
```

```python
pd.crosstab(index=copy["Automatic"], columns=copy["FuelType"], dropna=True)
```

```python
pd.crosstab(index=copy["Automatic"], columns=copy["FuelType"], dropna=True, normalize=True) #joint probability
```

```python
pd.crosstab(index=copy["Automatic"], columns=copy["FuelType"], dropna=True, normalize=True, margins=True) #marginal probability
```

```python
pd.crosstab(index=copy["Automatic"], columns=copy["FuelType"], dropna=True, normalize="index", margins=True) #conditional probability
```
