```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv("Toyota.csv", index_col=0, na_values=["??", "????"])
cars = cars.dropna(axis=0)
```

```python
sns.set(style="darkgrid")
sns.regplot(x=cars["Age"], y=cars["Price"])
```

```python
sns.regplot(x=cars["Age"], y=cars["Price"], fit_reg=False)
```

```python
sns.regplot(x=cars["Age"], y=cars["Price"], fit_reg=False, marker="*")
```

```python
sns.lmplot(x="Age", y="Price", data=cars, fit_reg=False, hue="FuelType", legend=True, palette="Set1")
```

```python
sns.displot(cars["Age"], kde=True)
```

```python
sns.displot(cars["Age"], kde=True, bins=5)
```

```python
sns.countplot(x="FuelType", data=cars)
```

```python
sns.countplot(x="FuelType", data=cars, hue="Automatic")
```

```python
sns.boxplot(y=cars["Price"])
```

```python
sns.boxplot(x="FuelType", y="Price", data=cars)
```

```python
sns.boxplot(x="FuelType", y="Price", data=cars, hue="Automatic")
```

```python
f, (ax_box, ax_hist) = plt.subplots(2, gridspec_kw={"height_ratios":(.15, .85)})
sns.boxplot(x=cars["Price"], ax=ax_box)
sns.histplot(cars["Price"], ax=ax_hist, kde=False)
```

```python
sns.pairplot(cars, kind="scatter", hue="FuelType")
```
