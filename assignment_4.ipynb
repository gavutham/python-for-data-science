```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as mse, accuracy_score, confusion_matrix, root_mean_squared_error as rmse
```

```python
data = pd.read_csv("ServiceTrain.csv")
```

```python
data["Service"] = [1 if x=="Yes" else 0 for x in data["Service"]]
```

```python
x = data.drop(["Service"], axis=1)
y = data["Service"]
```

```python
trainX, testX, trainY, testY = tts(x, y, test_size=0.3, random_state=0)
```

```python
x['OilQual'].value_counts()
```

```python
logi = LogisticRegression()
logi.fit(trainX, trainY)
```

```python
pred = logi.predict(testX)
```

```python
confusion_matrix(testY, pred)
```

```python
accuracy_score(testY, pred)
```
